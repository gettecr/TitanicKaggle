```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanicfile = '../data/raw/train.csv'
```

```python
df = pd.read_csv(titanicfile)
```

```python
df['Title']=df['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())
```

```python
df['Title'].head()
```

```python
df["Female"]=np.where(df['Sex']=='female',1,0)
```

```python
df.groupby('Title')['Age'].describe()
```

```python
df[df['Embarked'].isnull()]
```

```python
means = df.groupby('Title', as_index=False).agg(
                      {'Age':['mean','std']})
```

```python
means.columns=['Title','Age_mean',"Age_std"]
means=means.set_index('Title')
```

```python
means
```

```python
conditions=[(df['Title']=='Miss'),(df['Title']=='Mrs'),
           (df['Title']=='Mr'),(df['Title']=='Master')]
```

```python
Miss_ages = df.loc[(df["Title"]=='Miss'),'Age'].tolist()
Miss_ages = [x for x in Miss_ages if str(x) != 'nan']
```

```python
Mrs_ages = df.loc[(df["Title"]=='Mrs'),'Age'].tolist()
Mrs_ages = [x for x in Mrs_ages if str(x) != 'nan']
```

```python
Mr_ages = df.loc[(df["Title"]=='Mr'),'Age'].tolist()
Mr_ages = [x for x in Mr_ages if str(x) != 'nan']
```

```python
Master_ages = df.loc[(df["Title"]=='Master'),'Age'].tolist()
Master_ages = [x for x in Master_ages if str(x) != 'nan']
```

```python
df['Age_rand']=df['Age']
```

```python
df.loc[conditions[0],'Age_rand']=df.loc[conditions[0],'Age_rand'].apply(lambda x: np.random.choice( np.array(Miss_ages)))
```

```python
df.loc[conditions[1],'Age_rand']=df.loc[conditions[1],'Age_rand'].apply(lambda x: np.random.choice( np.array(Mrs_ages)))
```

```python
df.loc[conditions[2],'Age_rand']=df.loc[conditions[2],'Age_rand'].apply(lambda x: np.random.choice( np.array(Mr_ages)))
```

```python
df.loc[conditions[3],'Age_rand']=df.loc[conditions[3],'Age_rand'].apply(lambda x: np.random.choice( np.array(Master_ages)))
```

```python
df.groupby('Title')['Age_rand'].describe()
```

```python
df[df['Age_rand'].isnull()]
```

```python
df.loc[df['PassengerId']==767,"Age_rand"]=means.loc['Dr','Age_mean']
```

```python
df['Age_mod']=np.where(df['Age'].isnull(),df['Age_rand'],df['Age'])
```

```python
df[['Age','Age_mod','Age_rand']].head()
```

```python
df['Child']=np.where(df["Age_mod"]<16.0,1,0)
```

```python
import statsmodels.api as sm
```

```python
predictors = ['Pclass','Female','Child']
X = df[predictors]
y = df['Survived']
    
logit = sm.Logit(y, X)
result = logit.fit()
```

```python
result.summary()
```
